## Conclusão 

A partir dos dados e dos gráficos é possível chegar a conclusão que o PIB do estado não tem uma relação tão forte com o acesso de tecnologias, mas sim com a taxa em que novas tecnologias são adotadas pelas classes baixas em relação com a média do país,como TVs de tela fina ou internet de banda larga, contudo devem ser levados em conta outras características dos estados, como no caso das redes 3G e 4G, que tiveram um uso muito maior em 2013 em estados com uma classificação baixa, muito provavelmente por características geográficas, como Tocantins, classificado como número 26, porém mais de 96% do acesso a internet por pessoas de baixa renda foi feito por redes móveis enquanto o maior PIB estadual, São Paulo, teve somente 38% no mesmo ano. Esse fato se deve muito provavelmente pela geografia de cada estado, assim não se pode associar diretamente o PIB do estado com o maior acesso ou uso de novas tecnologias.
